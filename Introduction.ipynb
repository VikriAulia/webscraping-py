# Web Scraping with Beutiful Soup

```python
import requests
import html5lib
from bs4 import BeautifulSoup
```

## test response raw data 

```python
response = requests.get("https://en.wikipedia.org/wiki/Fallacy")
print(response.content)
```

## Response with BeautifulSoap

```python
soup = BeautifulSoup(response.content, "html5lib")
print(soup.prettify())
```

## example get table of content from wiki 

```python
table_of_content = soup.find("ul", attrs = {"id": "mw-panel-toc-list"})
print(table_of_content.prettify())
```

### Only the list text

```python
list_text = table_of_content.findAll("div", attrs = {"class": "vector-toc-text"})
print(list_text)
```

### only show text and strip the space

```python
for i in list_text:
    value = i.text.strip()
    for char in value:
        if char.isnumeric():
            continue
        elif char == ".":
            print("", end="\t")
        else:
            print(char, end="")
    print()
```
